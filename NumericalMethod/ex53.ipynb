```python
import numpy as np
import matplotlib.pyplot as plt
```

1. (a) modify bisect function so taht, in addtion to the approximate solution, the number of iterations needed to ahcieve the indicated tolerance is also returend.
2. (b) Plot the function f(x) = x^3-100cos(x) on the interval [-1,4]
3. (c) Find the root of the function using a tolerance of 0.0001

```python
def bisect (f,a,b,tol):
    x=(a+b)/2
    iter = 0
    while np.abs(f(x))>tol :
        print('iteration {:d} a={:.5f} f(a)={:.5f}, b={:.5f} f(b)={:.5f},)x={:.5f} f(x)={:.5f}'.format(iter,a,f(a),b,f(b),x,f(x)))
        if f(a)*f(x) < 0:
            b=x
        elif f(b)*f(x) < 0 :
            a=x
        x=(a+b)/2
        iter+=1
    return x,iter
```

```python
def test_function(x):
    return x**3-100*np.cos(x)
a = -1
b = 4
n = 20
dx=(b-a)/n
x=np.arange(a,b+dx,dx)
y2=test_function(x)
plt.plot(x,y2)
plt.grid()
plt.xlabel('x-axis')
plt.ylabel('f(x)')
plt.title(' x**3-100*np.cos(x)')
plt.plot()
```

```python
sol,iter=bisect(test_function,a,b,0.00001)

print(f'solution ={test_function(sol)}, iteration = {iter}')
```
