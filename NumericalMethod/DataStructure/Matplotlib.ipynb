```python
import numpy as np
import matplotlib.pyplot as plt

def pw(x):
    if x<=0:
        y=x**2
    else :
        y=x+1
    return y

vpw= np.vectorize(pw)
a=-2
b=2
n=100
dx=(b-a)/n
x=np.arange(a,b+dx,dx)
print(x)
y=vpw(x)
plt.plot(x,y)
plt.grid()
plt.xlabel('x-axis')
plt.ylabel('f(x)')
plt.title('when x<=0 f(x)=x^2 ,else f(x)=x+1 ')
plt.plot()
```

```python
a = -2
b = 0 
n = 50
dx = (b-a)/n
x = np.arange(a,b+dx,dx)
x[n]=b
y=vpw(x)
plt.plot (x,y,'b')
a=0
b=2
n=50
dx=(b-a)/n
x=np.arange(a+dx,b+dx,dx)
y=vpw(x)
plt.plot(x,y,'b')
plt.grid()
plt.plot(0,0,'b.',markersize=11)
plt.plot(0,1,'b.',fillstyle='none',markersize=11)
plt.show()
```

```python
A = np.array([[1,2,3,4],[-2,4,-3,5],[-1,3,-3,4]])
print(f'A = ',A)
a23 = A[1,2]
print('The value in the second row, third column is ',a23)
m,n = np.shape(A)
print('rows = {}. cols = {}'.format(m,n))
Arow3 = A[2,:]
print('The third row is ',Arow3)
Acol2 = A[:,1]
print('The second column is ',Acol2)
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[-1,3],[2,-5]])
print('A*B = ',A*B)
print('A/B = ',A/B)
print('3*A= ', 3*A)
```

```python
A = np.array([[1,2,3],[-1,2,-3]])
B = np.array([[1,2],[0,4],[-3,2]])
C = np.dot(A,B)
print('A dot B = \n',C)
```

```python
A = np.array([[1,2,3],[-1,2,-3],[0,2,5]])
A_inv = np.linalg.inv(A)
print('A_Inverse = \n', A_inv)
print('AA_inv= ) \n',np.dot(A,A_inv).round(3))
```

```python
A = np.array([[1,1,1],[2,-3,1],[1,2,-3]])
B = np.array([[6],[-1],[-4]])
X = np.linalg.solve(A,B)
print('X=',X)
```
