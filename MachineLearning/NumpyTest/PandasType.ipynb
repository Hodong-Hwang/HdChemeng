## Pandas
Pandas is scietific pacakage. it is built on top of Numpy and offers several data structures.
and it makes tabular data anaylsis and preprocessing very convenient


```python
import pandas as pd
```

```python
data=[10,8,6]
s = pd.Series(data)
print(s)
```

```python
data = [[1,10],[1,8],[1,6]]
df = pd.DataFrame(data,columns=['id','Value'])
print(df)
```

```python
s2 = pd.Series([1,1,1])
df2 = pd.DataFrame({'id':s2,'value':s})
print(df2)
```

```python
print(df['id'])
print(df.id)
print(df[['id']])
df.index=[100,101,102]
print(df)
```

```python
print(df.loc[101]) # returns 2nd row as a series; can provide a list for multiple rows selection
print(df.iloc[1]) # integer location-based selection; same result as above
print(df.loc[101,'Value'])
print(df.iloc[1, 1]) # same as above
```

```python
df3 = df.copy()
df3.id = 2 # make all items in column 'id' as 2
df3.Value *= 4 # multiply all items in column 'value' by 4
print(df3)
```

```python
df3 = df.append(df2) # a new object is retuned unlike Pythons append function
print(df3)
```
